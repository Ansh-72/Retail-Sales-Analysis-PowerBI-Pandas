```python
data['order_date']=pd.to_datetime(data['order_date'],format='%m/%d/%Y')
data['ship_date']=pd.to_datetime(data['ship_date'],format='%m/%d/%Y')
```

```python
data.dtypes
```

```python
data['year']=data['order_date'].dt.year
data['year']
```

#### Total revenue

```python
total_revenue=data['sales'].sum()
```

```python
total_revenue
```

```python
data.to_csv('retail_sales_data.csv',index=False)
```

### Total Profit


```python
total_profit=data['profit'].sum()
```

```python
total_profit
```

### Profit Margin%


```python
profit_margin=total_profit/total_revenue*100
profit_margin
```

### Average Delivery Day


```python
delivery_day= (data['ship_date']-data['order_date']).dt.days
average_delivery_day= delivery_day.mean()
average_delivery_day
```

### Total order

```python
total_order=data['order_id'].unique()
```

```python
total_order
```

### Total quantity

```python
total_quantity=data['order_quantity'].sum()
total_quantity
```

### revenue by category

```python
revenue_by_category=data.groupby('product_category')['sales'].sum()
```

```python
revenue_by_category
```

```python
profit_by_category_year=data.groupby(['year' , 'product_category'])['profit'].sum()
profit_by_category_year
```

### Top 10 product

```python
top_10_products = (
    data.groupby('product_name')['sales']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)
top_10_products
```

### ship mode %


```python
ship_mode_distribution =data.groupby('ship_mode')['order_id'].nunique()
```

```python
ship_mode_distribution
```

```python
ship_mode_percent = (ship_mode_distribution / ship_mode_distribution.sum()) * 100
ship_mode_percent
```
